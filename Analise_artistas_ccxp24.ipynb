```python
import pandas as pd

artist_valey = pd.read_csv("./Artistas CCXP24 - Página3.csv", delimiter=",", quotechar="\"", encoding="UTF-8")
artist_valey
```

```python
artist_valey = artist_valey.drop(columns=["Unnamed: 3","Unnamed: 5"])
artist_valey.Temas.unique()
```

```python
# Teste do codigo do gemini
def split_column(df, column_name):
    df[column_name] = df[column_name].str.lower()
    return df[column_name].str.split(',', expand=True).stack().reset_index(level=1, drop=True).drop_duplicates().sort_values().reset_index(drop=True).to_frame(column_name)
```

```python
# Criando os novos DataFrames
df_temas = split_column(artist_valey, 'Temas')
df_temas
```

```python
df_produtos = split_column(artist_valey, 'Produtos')
df_produtos
```

```python
# Salvando para ler melhor
df_temas.to_csv("Temas.csv")
df_produtos.to_csv("Produtos.csv")
```

```python
# Tentando buscar agora os artistas com temas e produtos que eu gosto
meu_temas = pd.read_csv("./Temas_meu.csv", index_col=0).reset_index().drop(columns=["index"])
meu_temas.head()
```

```python
meu_produtos = pd.read_csv("./Produtos_meu.csv", index_col=0).reset_index().drop(columns=["index"])
meu_produtos.head()
```

```python
def find_tema_ou_produto(df, temas, produtos):
    busca = []
    for i, row in df.iterrows():
        for j, t in temas.iterrows():
            if t["Temas"] in row["Temas"].lower():
                busca.append(row[["Nome","Mesa","Link Catálogo"]])
                for j, p in produtos.iterrows():
                    if p["Produtos"] in row["Produtos"].lower():
                        busca.append(row[["Nome","Mesa","Link Catálogo"]])
    return busca
```

```python
dados = find_tema_ou_produto(artist_valey, meu_temas, meu_produtos)
dados
```

```python
# Transformando em Dataframe e limpando dados
catalogo = pd.DataFrame(dados)
catalogo = catalogo.sort_values(by=["Mesa","Nome"]).drop_duplicates().reset_index(drop=True)
catalogo
```

```python
catalogo.to_csv("Catalogo_ccxp24.csv")
```
